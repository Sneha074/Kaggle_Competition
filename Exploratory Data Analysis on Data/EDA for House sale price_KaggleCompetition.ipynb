### Checking for the missing values

```python
features_with_na=[features for features in dataset.columns if dataset[features].isnull().sum()>1]

for feature in features_with_na:
    print(feature, np.round(dataset[feature].isnull().mean(), 4),  ' % missing values')
```

### realtionship between missing values and sale price

```python
for feature in features_with_na:
    data = dataset.copy()
    
    # let's make a variable that indicates 1 if the observation was missing or zero otherwise
    data[feature] = np.where(data[feature].isnull(), 1, 0)
    
    # let's calculate the mean SalePrice where the information is missing or present
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

### finding whether the numerical data is discrete or continuous

### finding relationship between discrete features nd saleprice

### finding continuous features

```python
continuous_feature = [feature for feature in numerical_features if feature not in discrete_features and feature not in temporal_features+['Id']]
continuous_feature
```

### relationship betwwen continuous features nd saleprice

```python
for feature in continuous_feature:

    data = dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.show()
```

#### EDA PART 2

#### using logarithmic transformation

```python
for feature in continuous_feature:
     data = dataset.copy()
     if 0 in data[feature].unique():
         pass
     else:
         data[feature] = np.log(data[feature])
         data['SalePrice'] = np.log(data['SalePrice'])
         plt.scatter(data[feature], data['SalePrice'])
         plt.xlabel(feature)
         plt.ylabel('SalePrice')
         plt.title(feature)
         plt.show()

```

### Detecting Outliers

```python
for feature in continuous_feature:
    data=dataset.copy()
    if 0 in data[feature].unique():
        pass
    else:
        data[feature] = np.log(data[feature])
        data.boxplot(column= feature)
       
        plt.ylabel(feature)
        plt.show()
    
```
